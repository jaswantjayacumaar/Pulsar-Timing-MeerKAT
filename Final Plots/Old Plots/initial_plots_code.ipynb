```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_data(fil, analysis):
    if analysis == 'Bayes' :
        data = np.loadtxt(fil)
        mjd = data[:,0]
        nudot = data[:,3]*1e-15
        nudot_err = data[:,4]*1e-15
    if analysis == 'fitwaves':
        data = np.loadtxt(fil)
        mjd = data[:,0]
        nudot = data[:,1]
        nudot_err = data[:,2]
    return [mjd, nudot, nudot_err]
```

```python
data_bayes1 = read_data('nudot.jbo.bayes.asc', analysis='Bayes')
data_bayes2 = read_data('nudot.mk.bayes.asc', analysis='Bayes')
data_bayes3 = read_data('nudot.comb.bayes.asc', analysis='Bayes')
data_fitwaves = read_data('nudot.mk.fitwaves.txt', analysis='fitwaves')
```

```python
#plt.subplot(2,2,1)
#plt.errorbar(data_fitwaves[0], data_fitwaves[1], yerr=data_fitwaves[2], color='k', capsize=2)
upper_error = data_bayes1[1] + data_bayes1[2]
lower_error = data_bayes1[1] - data_bayes1[2]
plt.plot(data_bayes1[0], data_bayes1[1], color='green')
plt.fill_between(data_bayes1[0], upper_error, lower_error, color='green', alpha=0.3)

#plt.subplot(2,2,2)
#plt.errorbar(data_fitwaves[0], data_fitwaves[1], yerr=data_fitwaves[2], color='k', capsize=2)
upper_error = data_bayes2[1] + data_bayes2[2]
lower_error = data_bayes2[1] - data_bayes2[2]
plt.plot(data_bayes2[0], data_bayes2[1], color='red')
plt.fill_between(data_bayes2[0], upper_error, lower_error, color='red', alpha=0.3)

#plt.subplot(2,2,3)
#plt.errorbar(data_fitwaves[0], data_fitwaves[1], yerr=data_fitwaves[2], color='k', capsize=2)
upper_error = data_bayes3[1] + data_bayes3[2]
lower_error = data_bayes3[1] - data_bayes3[2]
plt.plot(data_bayes3[0], data_bayes3[1], color='blue')
plt.fill_between(data_bayes3[0], upper_error, lower_error, color='blue', alpha=0.3)

#plt.subplot(2,2,4)
plt.errorbar(data_fitwaves[0], data_fitwaves[1], yerr=data_fitwaves[2], color='k', capsize=2)
#upper_error = data_fitwaves[1] + data_fitwaves[2]
#lower_error = data_fitwaves[1] - data_fitwaves[2]
#plt.plot(data_fitwaves[0], data_fitwaves[1], color='green')
#plt.fill_between(data_fitwaves[0], upper_error, lower_error, color='green', alpha=0.3)

plt.savefig("final.pdf")
plt.show()
```
